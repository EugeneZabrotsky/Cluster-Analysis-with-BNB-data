# Here will be shown the process of feature engineering. Specifically, we will aggregate raw dataset, which is a list of transactions, 

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelBinarizer
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
print(os.listdir("../input/bnb-dataset"))

# Any results you write to the current directory are saved as output.
```

```python
df_init = pd.read_csv("../input/bnb-dataset/Dataset_BNB.txt", sep = ';', encoding = 'utf-8')
df_init['TRAN_DATE'] = pd.to_datetime(df_init['TRAN_DATE'], format= '%d/%m/%Y')
```

```python
mccs = pd.read_csv('../input/bnb-dataset/MCCs.csv', encoding = 'utf-8')
mccs.fillna('unknown', inplace = True)
mccs['Группа'] = mccs['Группа'].map({"Отели и мотели" : 'hotels',
                                    "Авиалинии, авиакомпании": 'airlines',
                                    'Автомобили и транспортные средств': 'transport_vehicles',
                                    'Аренда автомобилей': 'auto_rental',
                                    'Бизнес услуги': 'business_service',
                                    'Государственные услуги': 'state_service',
                                    'Коммунальные и кабельные услуги': 'utility_service',
                                    'Контрактные услуги': 'contract_service',
                                    'Личные услуги': 'personal_service',
                                    'Магазины одежды': 'clothes', 
                                    'Неизвестно': 'unknown',
                                    'Оптовые поставщики и производители': 'wholesale',
                                    'Поставщик услуг': 'service', 
                                    'Продажи по почте/телефону': 'mail_phone_sales',
                                    'Профессиональные услуги': 'professional_service', 
                                    'Развлечения': 'entertainment',
                                    'Различные магазины': 'other_shops',
                                    'Ремонтные услуги': 'repair_service', 
                                    'Розничные магазины': 'retail_shops',
                                    'Транспорт': 'transport',
                                    'Членские организации': 'membership_organizations'})
mcc_map = dict(mccs[['MCC', 'Группа']].values)
```

```python
df = df_init.copy()
df_init.head(5).T
```

```python
df.groupby('CARD_TYPE').agg({'IS_CREDIT_CARD': np.max})
```

```python
pd.options.display.max_rows = 200
df['TRAN_COUNTRY'] = df['TRAN_COUNTRY'].apply(str.upper).apply(lambda x: 'BLR' if x == 'BY' else x)\
                                        .apply(lambda x: 'USA' if x == 'US' else x)
```

```python
df['GENDER'] = df['GENDER'].map({"M": 1, "F": 0})
df["TRAN_TYPE"] = df['TRAN_TYPE'].map({"PMT": 0, "CASH": 1})
df['CARD_TYPE'] = df['CARD_TYPE'].map({'Visa Classic' : 3 , 
                                       'Belcard-Maestro' : 1,
                                       'Visa Electron' : 2,
                                       'Visa Gold' : 4,
                                       'Visa Platinum' :6,
                                       'MasterCard Gold' : 4,
                                       'Visa Infinite' : 5,
                                       'Maestro' : 2,
                                       'MasterCard Standard': 3
                                      })

df['WEEKDAY'] = df['TRAN_DATE'].apply(pd.datetime.weekday)
df['IS_WEEKEND'] = df['WEEKDAY'] >= 5
df['MONTHDAY'] = df['TRAN_DATE'].apply(lambda date: date.day)
```

```python
region_code = 0
region_map={}
for region in df['CLIENT_REGION']:
    if region not in region_map:
        region_map.update({region:region_code})
        region_code+=1
region_map
df['CLIENT_REGION'] = df['CLIENT_REGION'].map(region_map)
```

```python
plt.plot(range(1, 32), df.groupby('MONTHDAY').agg({'TRAN_AMOUNT_BYN': np.sum})['TRAN_AMOUNT_BYN'])
```

```python
df['AMOUNT_IF_FOREIGN'] = df['TRAN_AMOUNT_BYN'] * ((df['TRAN_COUNTRY'] != 'BLR') * 1)
df['AMOUNT_IF_BLR'] = df['TRAN_AMOUNT_BYN'] * ((df['TRAN_COUNTRY'] == 'BLR') * 1)

df['AMOUNT_IF_CREDIT'] = df['TRAN_AMOUNT_BYN'] * df['IS_CREDIT_CARD']
df['AMOUNT_IF_NOT_CREDIT'] = df['TRAN_AMOUNT_BYN'] * (df['IS_CREDIT_CARD'] * (-1) + 1)

df['AMOUNT_IF_WEEKEND'] = df['TRAN_AMOUNT_BYN'] * df['IS_WEEKEND'] 
df['AMOUNT_IF_NOT_WEEKEND'] = df['TRAN_AMOUNT_BYN'] * (df['IS_WEEKEND'] * (-1) + 1)
```

```python
df['MCC_CODE_GROUP'] = df['MCC_CODE'].map(mcc_map).fillna("unknown")
```

```python
lb = LabelBinarizer()
lb.fit(df['MCC_CODE_GROUP'])
```

```python
mcc_groups = lb.classes_
```

```python
mcc_binarized = lb.transform(df['MCC_CODE_GROUP'])
```

```python
df = pd.concat(
    [df,
    pd.DataFrame(mcc_binarized, columns= lb.classes_)
    ], axis = 1
)
```

```python
mcc_groups_amount = []
for group in mcc_groups:
    group_amount = 'AMOUNT_IF_' + group
    mcc_groups_amount.append(group_amount)
    df[group_amount] = df[group] * df['TRAN_AMOUNT_BYN']
```

```python
cl = pd.DataFrame({"CLIENT_ID": np.unique(df["CLIENT_ID"])})
```

```python
grouped_cl = df.groupby(['CLIENT_ID'])
def add_feature(feature_name, col, func, res = cl, grouped = grouped_cl):
    res[feature_name] = grouped.agg({col: func})[col].values
```

```python
add_feature('SUM_AMOUNT', 'TRAN_AMOUNT_BYN', np.sum)
add_feature('MEAN_AMOUNT', 'TRAN_AMOUNT_BYN', np.mean)
add_feature('MAX_CARD_LVL', 'CARD_TYPE', np.max)
add_feature('GENDER', 'GENDER', np.max)
add_feature('MAX_AMOUNT', 'TRAN_AMOUNT_BYN', np.max)
add_feature('NUM_TRANS', 'TRAN_AMOUNT_BYN', len)
add_feature('NUM_COUNTRIES', 'TRAN_COUNTRY', lambda x: len(np.unique(x)))
add_feature('NUM_FOREIGN_COUTRIES', 'TRAN_COUNTRY', lambda x: len(np.unique(x[x != 'BLR'])))
add_feature('CLIENT_REGION','CLIENT_REGION',np.max)

cl['IS_ONLY_FOREIGN'] = (cl['NUM_COUNTRIES'] - cl['NUM_FOREIGN_COUTRIES'] != 1) * 1

add_feature('SUM_AMOUNT_FOREIGN', 'AMOUNT_IF_FOREIGN', np.sum)
add_feature('SUM_AMOUNT_BLR', 'AMOUNT_IF_BLR', np.sum)

cl['SUM_AMOUNT_FOREIGN_PERCENTAGE'] = cl['SUM_AMOUNT_FOREIGN'] / cl['SUM_AMOUNT']
cl['SUM_AMOUNT_BLR_PERCENTAGE'] = cl['SUM_AMOUNT_BLR'] / cl['SUM_AMOUNT']

add_feature('NUM_CONTRACTS', 'CONTRACT_ID', lambda x: len(np.unique(x)))
add_feature('NUM_CARDS', 'CARD_ID', lambda x: len(np.unique(x)))
add_feature('HAS_CREDIT_CARD', 'IS_CREDIT_CARD', np.max)
add_feature('SUM_AMOUNT_CREDIT_CARD', 'AMOUNT_IF_CREDIT', np.sum)
add_feature('SUM_AMOUNT_NOT_CREDIT_CARD', 'AMOUNT_IF_NOT_CREDIT', np.sum)

cl['SUM_AMOUNT_CREDIT_CARD_PERCENTAGE'] = cl['SUM_AMOUNT_CREDIT_CARD'] / cl['SUM_AMOUNT']
cl['SUM_AMOUNT_NOT_CREDIT_CARD_PERCENTAGE'] = cl['SUM_AMOUNT_NOT_CREDIT_CARD'] / cl['SUM_AMOUNT']

add_feature('AGE', 'AGE', np.mean)
add_feature('STD_AMOUNT', 'TRAN_AMOUNT_BYN', np.std)

cl['STD_AMOUNT_NORMALIZED'] = cl['STD_AMOUNT'] / cl['MEAN_AMOUNT']

add_feature('SUM_AMOUNT_WEEKEND', 'AMOUNT_IF_WEEKEND', np.sum)
add_feature('SUM_AMOUNT_NOT_WEEKEND', 'AMOUNT_IF_NOT_WEEKEND', np.sum)

cl['SUM_AMOUNT_WEEKEND_PERCENTAGE'] = cl['SUM_AMOUNT_WEEKEND'] / cl['SUM_AMOUNT']
cl['SUM_AMOUNT_NOT_WEEKEND_PERCENTAGE'] = cl['SUM_AMOUNT_NOT_WEEKEND'] / cl['SUM_AMOUNT']

add_feature('IS_RESIDENT', 'IS_RESIDENT', np.max)
add_feature('IS_CITIZEN', 'IS_CITIZEN', np.max)
```

```python
for group in mcc_groups_amount:
    add_feature('SUM_' + group, group, np.sum)
    cl['SUM_' + group + '_PERCENTAGE'] = cl['SUM_' + group] / cl['SUM_AMOUNT']
```

```python
cl.fillna(0, inplace = True)
```
